```python
# Step 1: Upload file from local system
from google.colab import files
uploaded = files.upload()

# Step 2: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 3: Load dataset (make sure file name matches exactly)
data = pd.read_csv("iris (2).csv")

# Step 4: Show dataset
print("First 5 rows:")
print(data.head())

# Step 5: Remove categorical column (species)
if 'species' in data.columns:
    X = data.drop('species', axis=1)
else:
    X = data

# Step 6: Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 7: Apply K-Means
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)

# Step 8: Get results
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("\nCluster Labels:\n", labels)
print("\nCentroids:\n", centroids)

# Step 9: Add cluster column
data['Cluster'] = labels

# Step 10: Visualization (using first 2 features)
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200)

plt.title("K-Means Clustering on Iris Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
